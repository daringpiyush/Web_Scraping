# Special-Sequences in Regular Expression:-

```python


# special sequences


# \d        -- matches any decimal digit --     [0-9]

regex = re.compile('\d')


# \D        -- matches any non-digit character  -- [^0-9]

regex = re.compile('\D')

# \s        -- matches any whitespace character

regex = re.compile('\s')

# \S        -- matches any non-whitespace character

regex = re.compile('\S')

# \w        -- matches any alphanumeric character -- [a-zA-Z0-9_]

regex = re.compile('\w')

# \W        -- matches any non-alphanumeric character -- [^ a-zA-Z0-9_]

regex = re.compile('\W')
```

```python
regex = re.compile('\d')
regex.match('0')
```

# Repeating Things:-

```python
# * character - this specifies that the previous character can be matched zero or more times, instead of exactly once.

regex = re.compile('[a-c]*')       # -- lower limit is 0 and the upper limit is infinity # No need of write a lot use * for this


print(regex.match('caaaaaaaaaaabcaaaaa'))   # any order no require a particular order.
```

```python
regex = re.compile('a*')       # -- lower limit is 0 and the upper limit is infinity # No need of write a lot use * for this


print(regex.match(''))       # if no 'a' then still gives a match
```

```python

# +  character -- this specifies that the previous character can be matched one or more times

# difference from * -- 0 - infinity         + 1 - infinity

regex = re.compile('a+')

print(regex.match('aaaaaaaaa'))  #if minimum 1 match then give match object only.

regex=re.compile('[ab]+')
print(regex.match(''))  # if minimum 1 match not there then not give match object 


# using character classes


regex = re.compile('[a-c]*')

print(regex.match(''))   # if none charcter then also shows match
```

```python
# ? question mark -- says the previous character can either come once or not at all

regex = re.compile('a?b')           # min - 0       max - 1
print(regex.match('ab'))             # match object
print(regex.match('aab'))            # not match coz range of a is 1 and given as 2


# {m,n}    m and n are integer values   -- This qualifier means there must be at least m repetitions, and at most n

regex = re.compile('a{2,4}')            # combinations are start from 2 to 4:-- aa aaa aaaa
print(regex.match('a'))                 # not match coz minimum range is 2
print(regex.match('aaaa'))            # match object
print(regex.match('aaaaaaa'))          # still match but only bcoz of start 4 char match there as span=(0, 4)



# * {0,}       # use this for max range is infinite

regex = re.compile('a{0,}')

print(regex.match('aaaaaaaa'))

# + {1,}   # use this for max range is infinite and min range as 1




# ? {0,1}
```

```python
# ^ character   -- says that the string should start with
regex = re.compile('^abc')
print(regex.match('a'))      #no match as there is compliment of abc require

# | character -- is the or operator

regex = re.compile('a|b')
print(regex.match('adcskdncks'))      # match object if a occur or b occur

# $ character -- matches the end of line

regex = re.compile('abc$')
print(regex.match('ab'))         # no match as this not end with c

print(regex.match('abc'))         # match as this end with abc
```

# Searching with Beautifulsoup:-

```python



# most popular methods

# find()
# find_all()        -- to keep it simple for now, it takes the tag name as parameter



# Kinds of filters which we can use to retrieve tags - filters sent as parameter to find/find_all methods

# string

print(soup.find_all('a'))
```

```python

# regular expression

# tag names start with b in the html

regex = re.compile('^b')

for tag in soup.find_all(regex):
    print(tag.name)
    
```

```python


# tag names contains t any where in the string

regex = re.compile('t')

for tag in soup.find_all(regex):
    print(tag.name)
 
```

```python
  # list

# all a and b tags in the html will give

for tag in soup.find_all(['p','a']):
    print(tag.name)
    
```

```python
# function for tag having class in itself

def has_class(tag):
    return tag.has_attr('class')

for tag in soup.find_all(has_class):
    print(tag.name)


```

# Find_all Functions:- It return all the tags or the output as per requirement

```python
# Signature: find_all(name, attrs, recursive, string, limit, **kwargs)


# name parameter:- it gives regex object in which the parameters are- string - True - function

a_tags = soup.find_all('a')
for a in a_tags:
    print(a)
```

```python
# attributes parameter:- Basically dicitonary

attr = {'class':'header-logo'}
first_a = soup.find_all(attrs=attr)
print(first_a)

#also use for multiple attributes aswell through key value pair in dict

```

```python
# limit parameter:- We can limit the output or filter it

a_tags = soup.find_all('a',limit=2)
print(a_tags)
```

```python
# string parameter:- if string match will find then it gives the output


regex = re.compile('gallery')

tag = soup.find_all(string=regex)
tag
```

```python
# **kwargs arguments:- when keyword arguments are not defined then we use **kwargs

tags = soup.find_all(class_='button')
for tag in tags:
    print(tag)
```

```python
# to write the class attribute of a tag - use       class_          because simple class is a keyword in Python


# recursive parameter :- find direct child of html i.e recursively parse whole tree


title = soup.find_all('title',recursive=False)

print(title)
                        # no output as inside no title found
```

```python
title = soup.find_all('title',recursive=True)

print(title)                                   #output as inside  title found
```

### Part:-02

```python

base_url = 'http://codingbat.com'

all_divs = soup.find_all('div',class_='summ')

all_links = [base_url + div.a['href'] for div in all_divs]

for link in all_links:
    inner_page = requests.get(link,headers={'user-agent':user_agent.chrome})
    inner_soup = BeautifulSoup(inner_page.content,'lxml')
    div = inner_soup.find('div',class_='tabc')
    question_links = [base_url + td.a['href'] for td in div.table.find_all('td')]
    
```

### Part:-03

```python
question_links = [base_url + td.a['href'] for td in div.table.find_all('td')]
for question_link in question_links:
    final_page = requests.get(question_link)
    final_soup = BeautifulSoup(final_page.content, 'lxml')
    indent_div = final_soup.find('div', attrs={'class':'indent'})
    problem_statement = indent_div.table.div.string
    siblings_of_statement = indent_div.table.div.next_siblings

    examples = [sibling for sibling in siblings_of_statement if sibling.string is not None]
    
    print(problem_statement)
    for example in examples:
        print(example)

    print('\n\n\n')
```

```python
# part 2

for link in all_links:
    inner_page = requests.get(link,headers={'user-agent':user_agent.chrome})
    inner_soup = BeautifulSoup(inner_page.content,'lxml')
    div = inner_soup.find('div',class_='tabc')
    question_links = [base_url + td.a['href'] for td in div.table.find_all('td')]


# part 3

    for question_link in question_links:
        final_page = requests.get(question_link)
        final_soup = BeautifulSoup(final_page.content, 'lxml')
        indent_div = final_soup.find('div', attrs={'class':'indent'})
        problem_statement = indent_div.table.div.string


        siblings_of_statement = indent_div.table.div.next_siblings

        examples = [sibling for sibling in siblings_of_statement if sibling.string is not None]

        print(problem_statement)
        for example in examples:
            print(example)

        print('\n\n\n')
```
